```python
import sympy as sp
import numpy as np
from sympy.plotting import plot,plot3d
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
t,s = sp.symbols('t , s')
```

```python

y1=sp.sin(t)*sp.Heaviside(t)
y2=sp.Heaviside(t)-sp.Heaviside(t-2)
#plot(y1,y2,(t,-1,7))
Y1=sp.laplace_transform(y1,t,s, noconds=True)
Y2=sp.laplace_transform(y2,t,s, noconds=True)
Y1
Y2
y1st=sp.inverse_laplace_transform(s*Y1, s, t)
y1st
y2st=sp.inverse_laplace_transform(s*Y2, s, t)
y2st
#plot(y1,y1st,(t,-1,7))
plot(y2,y2st,(t,-1,7))
```

```python
x=sp.Heaviside(t)-sp.Heaviside(t-2)
g=3/2*(sp.Heaviside(t)-sp.Heaviside(t-2))
X=sp.laplace_transform(x,t,s, noconds=True)
G=sp.laplace_transform(g,t,s, noconds=True)
X
G
y=sp.inverse_laplace_transform(X*G,s,t)
y
plot(y,(t,-1,4))
```
