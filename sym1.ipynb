```python
import sympy as sp
import numpy as np
from sympy.plotting import plot,plot3d
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

Definition der beiden Variablen

```python
t, s = sp.symbols('t, s')
y1 = sp.sin(t)*sp.Heaviside(t)
y2 = sp.Heaviside(t) - sp.Heaviside(t-2)
```

Ableitung berechnen

```python
Y1=sp.laplace_transform(y1, t, s, noconds=True)
Y2=sp.laplace_transform(y2, t, s, noconds=True)

y1ab=sp.inverse_laplace_transform(s*Y1, s, t)
y2ab=sp.inverse_laplace_transform(s*Y2, s, t)
```

```python
y1ab
y2ab
plot(y1ab,(t,-2,2*sp.pi))
plot(y2ab,(t,-2,2))
```
