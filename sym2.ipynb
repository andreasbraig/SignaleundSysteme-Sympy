```python
import sympy as sp
import numpy as np
from sympy.plotting import plot
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

Definition der beiden Variablen

```python
t, s = sp.symbols('t, s')
g = 3/2*(sp.Heaviside(t) - sp.Heaviside(t-2))
x = sp.Heaviside(t) - sp.Heaviside(t-2)
```

```python
G=sp.laplace_transform(g, t, s, noconds=True)
X=sp.laplace_transform(x, t, s, noconds=True)

y=sp.inverse_laplace_transform(G*X, s, t)
y
plot(y,(t,-2,6))
```

```python
g_f = sp.lambdify((t), g, "numpy")
step = 0.01
t_n = np.arange(-6,6,step)
g_n = g_f(t_n)

x_f = sp.lambdify((t), x, "numpy")
x_n = x_f(t_n)

y_n=np.convolve(g_n,x_n,"same")*step
plt.plot(t_n,y_n)
plt.grid()
```
